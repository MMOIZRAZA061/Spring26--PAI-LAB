```python
jug1 = 4
jug2 = 3
goal = 2

visited = set()

def dfs(x, y, path):
    if (x, y) in visited:
        return False

    visited.add((x, y))
    path.append((x, y))

    print("State:", (x, y))

    if x == goal or y == goal:
        print("Goal Reached")
        print("Path:", path)
        return True

    states = [
        (jug1, y, "Fill Jug 1"),
        (x, jug2, "Fill Jug 2"),
        (0, y, "Empty Jug 1"),
        (x, 0, "Empty Jug 2"),
        (x - min(x, jug2 - y), y + min(x, jug2 - y), "Pour Jug 1 -> Jug 2"),
        (x + min(y, jug1 - x), y - min(y, jug1 - x), "Pour Jug 2 -> Jug 1")
    ]

    for nx, ny, rule in states:
        if (nx, ny) not in visited:
            print("Rule Applied:", rule)
            if dfs(nx, ny, path):
                return True

    path.pop()
    return False

dfs(0, 0, [])
```
