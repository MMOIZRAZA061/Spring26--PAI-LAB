```python
def n_queens(n):
    solutions = []
    
    def place(row, board):
        if row == n:
            solutions.append(board[:])
            return
        
        for col in range(n):
            safe = True
            for r, c in enumerate(board):
                if col == c or abs(row - r) == abs(col - c):
                    safe = False
                    break
            
            if safe:
                board.append(col)
                place(row + 1, board)
                board.pop()
    
    place(0, [])
    return solutions


solutions = n_queens(4)
print("Number of solutions:", len(solutions))
for sol in solutions:
    print(sol)    
```
